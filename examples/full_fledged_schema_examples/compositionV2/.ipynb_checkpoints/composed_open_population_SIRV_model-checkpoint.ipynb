# Model A: Open population SIR model

```julia
sir = @stock_and_flow begin

    :stocks
    S
    I 
    R 

    :parameters
    rbirth
    cbeta
    rrecovery
    rdeath

    :dynamic_variables
    v_birth = rbirth * N

    v_inf₁ = cbeta * S
    v_inf₂ = v_inf₁ * I
    v_inf₃ = v_inf₂ / N

    v_rec = rrecovery * I
    v_deathS = S * rdeath
    v_deathI = I * rdeath
    v_deathR = R * rdeath

    :flows
    CLOUD => f_birth(v_birth) => S 
    S => f_inf(v_inf₃) => I 
    I => f_rec(v_rec) => R
    S => f_deathS(v_deathS) => CLOUD
    I => f_deathI(v_deathI) => CLOUD
    R => f_deathR(v_deathR) => CLOUD
    

    :sums
    N = [S, I, R]

end
```

```julia
Graph(sir)
```

# Model B: Open population SVI model

```julia
svi = @stock_and_flow begin
    
    :stocks
    S
    V
    I

    :parameters
    rvaccine
    rdeath
    cbeta
    evaccine # 1.0 - evaccine
    # TODO: Deal with this so I don't need to create another parameter

    :dynamic_variables
    v_vacc = S * rvaccine
    v_deathV = V * rdeath
    v_evaccine_complement = 1.0 - evaccine
    v_infv₁ = cbeta * v_evaccine_complement
    v_infv₂ = v_infv₁ * V
    v_infv₃ = v_infv₂ * I
    v_infv₄ = v_infv₃ / N

    
    :flows
    S => f_vacc(v_vacc) => V
    V => f_deathV(v_deathV) => CLOUD
    V => f_infv(v_infv₄) => I

    :sums
    N = [S, V, I]

end
```

```julia
Graph(svi)
```

# Composition 

## Using UWD-algebra

```julia
# define the UWD-algebra
sirv_uwd = @relation (S,I) begin
    seir(S,I)
    svi(S,I)
end;
display_uwd(sirv_uwd)
```

```julia
# define a foot of the structured multicospan
footS=foot(:S, :N, :S=>:N)
Graph(footS;schema="C0")
     
```

```julia
# define a foot of the structured multicospan
footI=foot(:I, :N, :I=>:N)
Graph(footI;schema="C0")
     
```

```julia
# open sir and svi stock and flow diagram with the feet defined before
open_sir=Open(sir, footS, footI)
open_svi=Open(svi, footS, footI)
# Compose those two models according the UWD-algebra
open_sirv = oapply(sirv_uwd, [open_sir, open_svi])
# the composed stock and flow diagram is the apex of the composed open stock and flow diagram
sirv = apex(open_sirv)
```

```julia
Graph(sirv)
```

```julia
# define values of constant parameters
p = LVector(
    cbeta=0.1, rbirth=0.001, rdeath=0.001, rrecovery=0.05, # for model sir
    rvaccine=0.01, evaccine=0.3, evaccine_complement = 0.7 # for model svi
)
# define initial values for stocks
u0 = LVector(
    S=990.0, I=10.0, R=0.0, V=0.0
)
```

```julia
prob_sirv = ODEProblem(vectorfield(sirv),u0,(0.0,100.0),p);
sol_sirv = solve(prob_sirv,Tsit5(),abstol=1e-8);
plot(sol_sirv)
```
